```python
import pandas as pd

data = pd.DataFrame()

df = pd.read_json("http://rajapinnat.ymparisto.fi/api/Hydrologiarajapinta/1.0/odata/Virtaama?$top=392&$filter=year(Aika)%20eq%202019%20and%20Paikka/Tila_Id%20eq%201&$expand=Paikka")
data = data.append(df)
```

```python
data.drop("odata.metadata",1,inplace=True)
```

```python
data
```

```python
data2 = pd.DataFrame.from_dict({(i,j): data[i][j] 
                           for i in data.keys() 
                           for j in data[i].keys()},
                       orient='index')
```

```python
data2.reset_index(drop=True, inplace=True)
```

```python
data2.head()
```

```python
paikka_data = data2["Paikka"].to_dict()
```

```python
paikka_data = pd.DataFrame.from_dict(paikka_data, orient = "index")
```

```python
paikka_data
```

```python
paikka_data.columns
```

```python
paikka_data.drop(["Suure_Id", "Nro", "KoordYkPohj", 'KoordYkIta', 'KoordPkPohj', 'KoordPkIta', 'KoordLat',
       'KoordLong', 'KoordTarkkuus', 'Karttalehti', 'Omistaja_Id', 'Tila_Id',
       'Laitteisto_Id', 'H_Vesal_Id', 'VesalNimi', 'H_PaaVesal_Id',
       'PaaVesalNimi', "Jarvi_Id", "JarviNro", "JarviNimi", "H_Kunta_Id"],1, inplace=True)
```

```python
paikka_data
```

```python
# KoordErTm == epsg:3067

from pyproj import Proj, transform

inProj = Proj(init='epsg:3067')
outProj = Proj(init='epsg:4326')
x1,y1 = paikka_data["KoordErTmIta"].tolist(), paikka_data["KoordErTmPohj"].tolist()
x2,y2 = transform(inProj,outProj,x1,y1)
```

```python
paikka_data["lat"] = y2
```

```python
paikka_data["lon"] = x2
```

```python
import pandas as pd
import requests

paikka_data = pd.read_csv("virtaama.csv")

def update(havainto, kuntanimi, vuosi):
    paska = pd.DataFrame()
    r = requests.get("http://rajapinnat.ymparisto.fi/api/Hydrologiarajapinta/1.0/odata/{0}?$filter=Paikka/Tila_Id%20eq%201%20and%20Paikka/KuntaNimi%20eq%20%27{1}%27%20and%20year(Aika)%20eq%20{2}".format(havainto, kuntanimi, vuosi)).json()
    df = pd.DataFrame.from_dict(r)
    paska = paska.append(df, sort = True)
    
    r2 = requests.get("http://rajapinnat.ymparisto.fi/api/Hydrologiarajapinta/1.0/odata/{0}?$filter=Paikka/Tila_Id%20eq%201%20and%20Paikka/KuntaNimi%20eq%20%27{1}%27%20and%20year(Aika)%20eq%20{2}&$skip=500".format(havainto, kuntanimi, vuosi)).json()
    df2 = pd.DataFrame.from_dict(r2)
    paska = paska.append(df2, sort = True)
    
    r3 = requests.get("http://rajapinnat.ymparisto.fi/api/Hydrologiarajapinta/1.0/odata/{0}?$filter=Paikka/Tila_Id%20eq%201%20and%20Paikka/KuntaNimi%20eq%20%27{1}%27%20and%20year(Aika)%20eq%20{2}&$skip=1000".format(havainto, kuntanimi, vuosi)).json()
    df3 = pd.DataFrame.from_dict(r3)
    paska = paska.append(df3, sort = True)
    
    if "odata.metadata" and "odata.nextLink" in paska.columns:
        paska.drop(["odata.metadata", "odata.nextLink"],1, inplace = True)
    else:
        paska.drop(["odata.metadata"],1,inplace = True)
    
    paska.reset_index(inplace=True, drop=True)
    
    df2 = pd.DataFrame.from_dict({(i,j): paska[i][j] 
                            for i in paska.keys() 
                            for j in paska[i].keys()},
                        orient='index')
    
    df2.reset_index(inplace=True, drop=True)
    df2.drop(["ArvoMin", "ArvoMax", "Lippu_id", "huomautus"],1,inplace=True)
    df2.Aika = pd.to_datetime(df2.Aika)
    df2.to_csv("virtaama_mittaukset.csv", index = None, mode = "a", header = None)
    
```
